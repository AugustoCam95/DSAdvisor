```python
import pandas as pd 
from scipy.stats import chisquare
```

```python
df = pd.read_csv("mushrooms.csv")
```

```python
df=df.apply(lambda x : pd.factorize(x)[0])+1
```

```python
df
```

```python
pd.DataFrame([chisquare(df[x].values,f_exp=df.values.T,axis=1)[0] for x in df])
```
