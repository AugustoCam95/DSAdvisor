```python
import os 
import pandas as pd
```

```python
df = pd.read_csv("abalone.csv")
```

```python
data = df.describe()
```

```python
data
```

```python
std = data.iloc[[2]]
```

```python
mean = data.iloc[[1]]
```

```python
stdray = std.values
meanray = mean.values
```

```python
cv = []
for i in range(len(stdray)):
    cv.append(stdray[i]/meanray[i])
```

```python
cv[0]
```

```python
data.loc["cv"] = cv[0]
```

```python
data
```

```python
data = data.reindex(["count","mean","std","cv","min","25%","50%","75%","max"])
```
